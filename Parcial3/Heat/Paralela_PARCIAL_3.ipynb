```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('https://raw.githubusercontent.com/juanrod1028/ComputacionParalela/master/Parcial3/Heat/resultados.csv')
```

```python
datos
```

```python
plt.rcParams['figure.figsize']=[16,8]
plt.title("Comparativa de rendimiento entre python y cython")
plt.plot(datos["n_steps"],datos['tiempo_python'])
plt.plot(datos["n_steps"],datos['tiempo_cython'])
plt.legend(["python","cython"])
plt.xlabel("Número de pasos")
plt.ylabel("Tiempo (Seg)")
plt.grid()
plt.show()
```

```python
plt.rcParams['figure.figsize']=[16,8]
plt.title("Comparativa de rendimiento entre python y cython")
plt.plot(datos["n_steps"],datos['speedup'])
plt.legend(["python/cython"])
plt.xlabel("Número de pasos")
plt.ylabel("Tiempo (Seg)")
plt.grid()
plt.show()
```
